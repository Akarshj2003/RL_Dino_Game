```python
import os
from mss import mss#for capture screen
import pydirectinput#sent command
import cv2
import numpy as np
import pytesseract# for ocr in game
tesseract_path = os.getenv('TESSERACT_PATH', r'C:\Program Files\Tesseract-OCR')
pytesseract.pytesseract.tesseract_cmd = os.path.join(tesseract_path, 'tesseract.exe')
from matplotlib import pyplot as plt # to see frames
import time
from gym import Env#for envirnment
from gym.spaces import Box, Discrete
```

# Environment

## create

```python
class WGame(Env):
    def __init__(self):
        super().__init__()
        self.observation_space = Box(low=0,high=255,shape=(1,83,100),dtype=np.uint8)
        self.action_space = Discrete(3)
        self.cap = mss()
        self.game_location = {'top':150,'left':0,'width':900,'height':400}
        self.done_location = {'top':305,'left':430,'width':400,'height':70}
    def step(self, action):
        pass
    def render(self):
        pass
    def reset(self):
        pass
    def get_observation(self):
        raw =np.array(self.cap.grab(self.game_location))[:,:,:3]
        gray = cv2.cvtColor(raw, cv2.COLOR_BGR2GRAY)
        resize =cv2.resize(gray,(100,83))
        channel= np.reshape(resize,(1,83,100))
        return channel
        
    def get_done(self):
        done_cap = np.array(self.cap.grab(self.done_location))[:,:,:3] 
        done_strings=['GAME','GAHE']
        f=False
        res=pytesseract.image_to_string(done_cap)[:4]
        if res in done_strings:
            f=True
        return res,f,done_cap
        
    def close(self):
        pass
```

```python
en= WGame()
```

```python
text,done,done_cap=en.get_done()
```
